Необходимо собрать информацию по продуктам питания с сайта: Список протестированных продуктов на сайте Роскачество Приложение должно анализировать несколько страниц сайта (вводим через input или аргументы).

Получившийся список должен содержать:

    Наименование продукта.
    Все параметры (Безопасность, Натуральность, Пищевая ценность, Качество) Не забываем преобразовать к цифрам
    Общую оценку
    Сайт, откуда получена информация.

Общий результат можно вывести с помощью dataFrame через Pandas. Сохраните в json либо csv.

```python
import requests
from bs4 import BeautifulSoup as bs 
import json
import pandas as pd
```

```python
urls = []
urls.append('https://rskrf.ru/ratings/produkty-pitaniya/ryba-i-moreprodukty/ikra-lososevykh/')
urls.append('https://rskrf.ru/ratings/produkty-pitaniya/bakaleya/ris-kruglozernyy/')
```

```python
def ParsePage(url, goods):
    req = requests.get(url)
    if req.status_code == 200:
        soup = bs(req.content, 'html.parser')
        # все данные сложены уже в DataLayer, поэтому не будем парсить страницу, а 
        # только вытащим из скрипта переменную
        jstring = soup.prettify().split('var data = ', -1)[1].split(';', 1)[0]
        data = json.loads(jstring)
        items = data['items']
        for x in items:
            ind = { i['id']: i for i in x['indicator'] }
            try:
                # могут быть пропуски в категориях - поставим NA
                try:
                    marking = ind[1]['value']
                except:
                    marking = 'NA'
                    
                try:
                    microbio = ind[2]['value']
                except:
                    microbio = 'NA'
                    
                try:
                    unsafe = ind[3]['value']
                except:
                    unsafe = 'NA'
                
                try:
                    quality = ind[4]['value']
                except:
                    quality = 'NA'
                
                g = {
                    "name" : x['name'], 
                    "stars" : x['star'], 
                    "price" : x['price'],
                    "marking" : marking, 
                    "microbio" : microbio, 
                    "unsafe" : unsafe, 
                    "quality" : quality
                }
                goods.append(g)
            except:
                print(x['name'])
    else:
        print(req.status_code)
```

```python
goods = []
```

```python
for u in urls:
    ParsePage(u, goods)
```

```python
pd.set_option('display.max_rows', None)
df = pd.DataFrame(goods)
```
